## Aggregating Data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/bios.csv")
coffee = pd.read_csv("./data/coffee.csv")
```

```python
df.head()
```

```python
# Value counts is used for finding what values/categories are the most frequent ones
df["born_city"].value_counts()
```

```python
df[df["born_country"]=="USA"]["born_city"].value_counts()
```

```python
df.groupby("born_country")["height_cm"].mean()
```

```python
coffee.groupby("Coffee Type")["Units Sold"].sum()
```

```python
coffee.groupby("Coffee Type").agg({"Units Sold":"sum","Day":"count"}) 
```

```python
coffee.groupby(["Coffee Type","Day"]).agg({"Units Sold":"sum","Day":"count"}) 
```

```python
df = pd.DataFrame({
    "user_id": [1, 1, 1, 2, 2, 3],
    "product": ["A", "B", "A", "A", "C", "B"],
    "price": [10, 20, 15, 10, 30, 20],
    "quantity": [1, 2, 1, 3, 1, 2]
})
```

```python

print(df.groupby("user_id")["product"].value_counts())
print(df.groupby("user_id")["price"].count())
print(df.groupby("user_id")["price"].sum())
```

```python
print(df.groupby("user_id")["price"].max())
print(df.groupby("user_id")["price"].min())
```

```python
df.head()
```

```python

df.groupby("user_id")["price"].agg(["sum","mean","max","min"])
```

```python
df.groupby("user_id").agg(avg_price = ("price","mean"), product_quantity=("quantity", "sum"),counting=("product","unique"))
```

```python
df.groupby(["user_id","product"])["price"].sum()
```

```python
df.groupby("user_id")["price"].transform("sum")
```
