### Adding & Removing Columns

```python
import pandas as pd
```

```python
bio = pd.read_csv("./data/bios.csv")
bios = bio.copy()
```

```python
bios["first_name"] = bios["name"].str.split(' ').str[0]
```

```python
bios.head()
```

```python
bios["born_datetime"] = pd.to_datetime(bios["born_date"])
```

```python
bios.head()
```

```python
# As you can see born_datetime column that is just created has datetime64 Data type
bios.info()
```

```python
bios["year"] = bios["born_datetime"].dt.year
```

```python
bios[["first_name","year"]].head()
```

```python
bio.to_csv("./data/bios.csv",index = False)
# This will remove index values in the original saved csv just for users sake
```

```python
# Applying lambda function 
bios["height_category"] = bios["height_cm"].apply(lambda x : 'Short' if x < 145 else "Average" if x < 170 else "Tall")
```
