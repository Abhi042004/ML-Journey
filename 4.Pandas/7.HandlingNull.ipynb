### Handling Null Values in the data

```python
import pandas as pd
import numpy as np
```

```python
toffee = pd.read_csv("././1.Instructors_Tutorial/warmup-data/coffee.csv")
coffee = toffee.copy()
```

```python
coffee.head()
```

```python
coffee.loc[0:2, "Units Sold"] = np.nan
```

```python
coffee.isnull().sum()
```

```python
coffee[coffee["Units Sold"].isnull()]
```

```python
coffee.fillna(coffee["Units Sold"].mean())
```

```python
coffee = coffee.dropna()
coffee.isnull().sum()
```
