### Filtering Data

```python
import pandas as pd
```

```python
df = pd.read_excel("1.Instructors_Tutorial\data\olympics-data.xlsx")
```

```python
df.head()
```

```python
# Conditions can be applied it is just like masking numpy
df[df["weight_kg"]>170]
```

```python
# Multiple conditions can be applied by using ( ) & ( ) - like this
df[(df["height_cm"] == 150) & (df["weight_kg"]>50)].head()
```

```python
# By Using .loc you can make this simpler
df.loc[(df["height_cm"]>214) & (df["born_country"] == "USA")  , ["name" , "height_cm"]].head()
```

```python
df.loc[df["name"].str.contains("Abhi")]
```

```python
df.loc[df["name"].str.contains("Keith|Abhi", case = False)].head()
# case False means that keith Keith KEITH all are same
```

```python
df.loc[(df["born_country"].isin(["USA","CAN","GBR"])) & (df["name"].str.startswith("Keith")), ["name"]].head()
```
