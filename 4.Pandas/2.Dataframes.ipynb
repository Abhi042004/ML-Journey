### Accessing Data with Pandas

```python
# Loc is used for accessing elements by rows and columns
# Selecting single row
coffee.loc[[0]]
```

```python
# To select multiple elements
coffee.loc[[0,4,2]]
```

```python
# Slicing to get elements
coffee.loc[8:14]
```

```python
# Selecting a single column
# coffee["Day"]
# coffee.loc[ : ,'Day']
coffee.Day
```

```python
# To get a single element using row and column
coffee.loc[3,'Day']
```

```python
# To get multiple rows and specific columns
coffee.loc[5:11, ["Day","Coffee Type"]]
```

```python
# Using slicing to get values
coffee.loc[6:9 , "Coffee Type":"Units Sold"]
```

```python
# Similarly you can do all these operations for Iloc
# Iloc is basically accessing elements using index like in arrays unlike in Loc
# In loc it is accessed by name of the index and the column
```
