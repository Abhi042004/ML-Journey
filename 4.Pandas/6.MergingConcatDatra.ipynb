### Merging & Concatenating Data

```python
import pandas as pd
```

```python
df = pd.read_csv("./1.Instructors_Tutorial/data/bios.csv")
bios = df.copy();
noc = pd.read_csv("./1.Instructors_Tutorial/data/noc_regions.csv")
```

```python
bios.head()
```

```python

noc.head()
```

```python
# So what we did basically is merged two different datasets left_on is the column 
# which is present in bios and right_one is in noc one and we are merging two of them
# how by inner join meaning the values which are common in both the selected columns
# will only be there

# for example there is left_inner join it joins basis of what values are present in the
# left column/dataset if it has row 1 2 4 but right one has 1 2 3 
# so the resultant row will be 1 2 4 and 4th row will be having null column values 
# will ignore any extra rows in right one

# simmillarly right_inner join
# outer join is there which just combines all the values of both the columns 
# regardless of whether they are present in left, right column or not
# if there are not present null will be filled
bios = pd.merge(bios, noc, left_on="born_country", right_on="NOC", how="inner")
```

```python
bios.head(5)
```

```python
bios=bios.drop(["born_country","notes"], axis=1)
bios.head(4)
```

```python
bios = bios.rename(columns={"NOC_x":"Country", "NOC_y":"Country_Code"})
bios.head(4)
```

```python
# these get the rows based on these two column where values are not equal
bios[(bios["Country"]) != bios["region"]][["name","Country","region"]].head()
```

```python
usa = bios[bios["Country_Code"]=="FRA"].copy()
gbr = bios[bios["region"]=="Germany"].copy()
```

```python
new_df = pd.concat([usa,gbr])
```
