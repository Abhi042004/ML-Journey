### Adding and Removing Columns

```python
import pandas as pd
```

```python
coffee = pd.read_csv("./data/coffee.csv")
```

```python
# Write now you can see toffee is is also using same data as coffee
# Meaning if you change toffee or coffee both will change
# coffee["Price"] = 4.99
toffee = coffee
```

```python
# so we do toffee.copy() to create a fresh copy and now both are independent
toffee = coffee.copy()
```

```python
# You can create a new column like this
coffee["Price"] = 4.99
```

```python
coffee.head()
```

```python
import numpy as np
coffee["New_Prices"] = np.where(coffee["Coffee Type"] == "Espresso",3.99,5.99)
coffee.head()
```

```python
# To drop a row
coffee.drop(0)
```

```python
# To drop a column there are two options
# coffee.drop("Price" ,axis = 1)
# if you add inplace = True then it will
# permanently delete the row or column from the data
coffee.drop(columns = "Price")
```

```python
toffee
```

```python
coffee = coffee.drop(columns="Price")
```

```python
coffee["Revenue"] = coffee["Units Sold"]*coffee["New_Prices"]
```

```python
coffee.head()
```

```python
coffee.rename(columns={"New_Prices":"Prices"},inplace=True)
```
