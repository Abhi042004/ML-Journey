```python
import pandas as pd
```

```python
# A DataFrame is a table with row, columns and labels
# While Creating a DataFrame -  You can also specify column and indexes name also
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=["A","B","C"])
df
# similarly you can do for rows also by writing index like we wrote in columns
```

```python
# To view first few rows use head()
print(df.head(),"\n")
# To view some specified number of rows
print(df.head(2),"\n")
# To view last few rows
print(df.tail(),"\n")
# To view last few specified number of rows
print(df.tail(1),"\n")
```

```python
# Columns
print(df.columns.tolist())
# Index
print(df.index.tolist())
```

```python
print(df.info(),"\n")
print(df.describe())
```

```python
# Gives count of number of unique values in the each column
print(df.nunique())
# Gives the unique values in a column
print(df["B"].unique())
# Gives number of elements in the dataframe
print(df.size)
# Gives number of rows and columns in the dataframe
print(df.shape)
```
