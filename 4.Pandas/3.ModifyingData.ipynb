```python
# Can also modify values
coffee.loc[1,"Units Sold"] = 10
# Can also modify multiple values using slicing
coffee.head()
```

```python
# sort_values sorts in ascending order by default
# syntax: dataframe.sort_values("column", ascending = True(by default)/False)
coffee.sort_values("Units Sold", ascending=False)
```

```python
# Simple rule: Place bracket for multiple values
coffee.sort_values(["Units Sold","Coffee Type"], ascending=[False,True])
```
