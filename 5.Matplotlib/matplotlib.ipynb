## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Creating a line plot
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Label")
plt.title("Basic Line Plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

#Customized Lineplot
plt.plot(x,y,color="green",linestyle = "-.",marker='o',markersize = 8,linewidth = 3)
plt.grid()
plt.show()
```

```python
import pandas as pd
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [2000, 2500, 3000, 2800, 3500, 4000],
    "Expenses": [1500, 1800, 2000, 1900, 2200, 2500]
}

df = pd.DataFrame(data)
```

```python
# Creating multiple plots in a single figure

# Total Size of the figure
plt.figure(figsize = (9,8))

# plt.subplot(row,column,which position)
plt.subplot(2,2,1) 
plt.plot(df["Month"],df["Expenses"],color = "red")


plt.subplot(2,2,2)
plt.plot(df["Month"],df["Revenue"],color = "blue")


plt.subplot(2,2,3)
plt.plot(df["Month"],df["Revenue"],color = "green")
plt.show()


```

```python
# Barplot
products = ["Laptop", "Tablet", "Phone", "Monitor", "Keyboard"]
units_sold = [50, 35, 80, 40, 60]
```

```python
# Creating BarPlot

plt.bar(products,units_sold,color="red",width=0.5, edgecolor="black",linewidth=2,label="Units_Sold")
plt.xlabel("Products")
plt.ylabel("Units_Sold")
plt.title("Sold Products with Quantity")
plt.legend()
plt.show()
```

```python
# Histogram used to represent distribution of the data
scores = [45, 50, 55, 60, 65, 70, 72, 75, 78, 80,
          82, 85, 88, 90, 92, 95, 98, 100]
```

```python
plt.hist(scores,bins=5,color="green",edgecolor = "black")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot

hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [45, 50, 60, 65, 70, 78, 85, 90]
```

```python
plt.scatter(hours_studied,exam_scores, color = "blue", marker="+")
plt.show()
```

```python
# PI Chart
products = ["Laptop", "Tablet", "Phone", "Monitor", "Keyboard"]
units_sold = [50, 35, 80, 40, 60]
colors = ["#4E79A7", "#59A14F", "#F28E2B", "#E15759", "#76B7B2"]
explode = [0, 0, 0.4, 0, 0]
```

```python
plt.pie(units_sold,labels=products,colors=colors,autopct="%1.1f%%",explode = explode,shadow = True )

plt.show()
```
