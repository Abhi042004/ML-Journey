```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/krishnaik06/Complete-Python-Bootcamp/refs/heads/main/10-Data%20Analysis%20With%20Python/sales_data.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
product_by_price =  df.groupby("Product Category" ,as_index=False)["Payment Method"].value_counts()
plt.bar(product_by_price["Product Category"],product_by_price["count"],color="blue")
```

```python
trend = df.groupby("Date")["Total Revenue"].sum().reset_index()
plt.plot(trend["Date"],trend["Total Revenue"],color = "blue")
# plt.bar(trend["Date"],trend["Total Revenue"],color = "blue")
plt.show()
```
