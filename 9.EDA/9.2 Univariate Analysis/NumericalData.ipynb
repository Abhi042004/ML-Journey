## EDA on Numerical Data

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./train.csv')
```

```python
df.head(5)
```

### 1.Histogram

```python
sns.histplot(data = df, x = 'Age', bins = 8)
```

```python
ax = sns.histplot(data=df, x="Age", bins=8)

for container in ax.containers:
    ax.bar_label(container)
```

### 2.Distplot

```python
### Distplot is depericated 
### Default value of histplot stat parameter is count stat = count to get density you can change it to density
# sns.histplot(data = df, x = 'Age', bins=8, stat="density", kde=True)
sns.kdeplot(data = df, x = 'Age')
```

### 3.Boxplot

```python
sns.boxplot(data=df, x = 'Age')
```

```python
print(df['Age'].min())
print(df['Age'].max())
print(df['Age'].mean())
```
