## Exploratory Data Analysis

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("./train.csv")
```

```python
df.sample(5)
```

### Categorical Data

### 1.Countplot

```python
sns.countplot(data=df, x='Survived')
```

```python
sns.countplot(data = df, x = 'Pclass')
```

```python
sns.countplot(data = df, x = 'Sex')
```

```python
sns.countplot(data = df, x = 'Embarked')
```

```python
df.value_counts(df['Survived']).plot(kind = 'bar')
print(df.value_counts(df['Survived']))
```

### 2.Pie Chart

```python
df.value_counts(df['Survived']).plot(kind = 'pie', autopct = '%.2f%%')
```

```python
df.value_counts(df['Sex']).plot(kind = "pie", autopct= '%.2f%%')
```
