```python
all_rows = []
len = 0

cursor = None

while(True):

    len +=1
    print(len)

    if len == 900:
        break;

    params = {
        "country": "IN"
    }

    if cursor:
        params["cursor"] = cursor

    response = requests.get(url,headers=headers,params=params)
    data = response.json()
    
    page_df = pd.DataFrame(data["shows"])
    page_df= page_df[page_df['showType']=='movie']
    page_df["year"] = page_df.get("releaseYear") 
    page_df = page_df[['id','title','year','overview']]
    

    all_rows.append(page_df)

    if data.get("hasMore") == False:
        break

    cursor = data.get("nextCursor")
    
    time.sleep(0.5)
```

```python
df = pd.concat(all_rows,ignore_index=True)
df.head()
```
