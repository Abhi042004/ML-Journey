## Principle Component Analysis

```python
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data

# Step 1: Standardize data
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_standardized = (X - mean) / std

# Step 2: Compute covariance matrix
cov_matrix = np.cov(X_standardized.T)

# Step 3: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort eigenvalues in descending order
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Step 5: Select top 2 principal components
W = eigenvectors[:, :2]

# Step 6: Project data onto principal components
X_pca_manual = X_standardized @ W

print("Manual PCA Result Shape:", X_pca_manual.shape)
print(X_pca_manual[:5])
```

### PCA using sklearn

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
iris = load_iris()
X = iris.data

# Step 1: Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Sklearn PCA Result Shape:", X_pca.shape)
print(X_pca[:5])
```
