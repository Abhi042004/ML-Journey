## Working With Dates

### Extracting Dates

```python
date['date_year'] = date['date'].dt.year
date.head()
```

### Extracting Months

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

### Extracting Months_Name

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

### Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.sample(5)
```

```python
# is weekend?
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
date.head()
```

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
date['date_quarter'] = date['date'].dt.quarter
date.head()
```

```python
date['date_semester'] = np.where(date['date_quarter'].isin([1,2]),1,2)
date.head()
```

### Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
np.round((today - date['date']) / np.timedelta64(30, 'D'), 0)
```
