### OneHotEncoding using pandas

```python
pd.get_dummies(df, columns=['fuel','owner'])
```

### (n-1) OneHotEncoding using Pandas

```python
pd.get_dummies(df, columns=['fuel','owner'], drop_first=True)
```

#### Problem with this is this method does'nt remember the position of the column, it changes everytime

### OneHotEncoding using scikit-learns

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, :4], df.iloc[:, -1], random_state=42, test_size=0.3)

x_train.shape,y_train.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)

x_train_new = oe.fit_transform(x_train[['fuel','owner']])
x_test_new = oe.transform(x_test[['fuel','owner']])
```

```python
x_test_new
```

```python
np.hstack((x_train[['brand','km_driven']].to_numpy(),x_train_new))
```

### OneHotEncoding on Brands

```python
counts = df['brand'].value_counts()
df['brand'].nunique()
```

```python
threshold = 100
repl = counts[counts <= threshold].index
repl
```

```python
pd.get_dummies(df['brand'].replace(repl, 'others')).astype(int).sample(5)
```
