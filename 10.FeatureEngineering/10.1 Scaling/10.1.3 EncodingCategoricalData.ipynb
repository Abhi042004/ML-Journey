## Encoding

### Categorical Data is divided into two types - nominal(no comparision) and ordinal(has a order)  
### When our input cols have a Categorical Data(Ordinal) we do ordinal encoding
### When our input cols have a Nominal Categorical Data we do One-Hot Encoding
### When our output col have a Categorical Data(ordinal) we do label encoding

```python
import pandas as pd
```

```python
df = pd.read_csv('./Dataset/Customer.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,2:-1],df.iloc[:, -1],random_state=42, test_size= 0.2)
x_train.shape, x_test.shape
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ode = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
ode.fit(x_train)

x_train = ode.transform(x_train)
x_test = ode.transform(x_test)
```

```python
ode.categories
```

```python
x_train = pd.DataFrame(x_train, columns=[['review','education']])
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(y_train)

y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
